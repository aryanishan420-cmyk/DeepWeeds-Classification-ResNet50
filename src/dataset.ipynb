```python
import pandas as pd
import os
import shutil
from tqdm import tqdm

# file locations
labels_file = "../data/raw/labels/labels.csv"
images_folder = "../data/raw/images"
output_folder = "../data/processed"

# read the label file
data = pd.read_csv(labels_file)

print("Columns in file:", data.columns)
data.head()

# loop through each row and copy images into their species folder
for i, row in tqdm(data.iterrows(), total=len(data)):

    # get image name
    img_name = str(row["Filename"])

    # some filenames don’t have extension, so add .jpg if missing
    if not img_name.lower().endswith(".jpg"):
        img_name = img_name + ".jpg"

    # species will be used as folder name
    species = row["Species"]

    # full paths
    src_path = os.path.join(images_folder, img_name)
    dest_folder = os.path.join(output_folder, species)

    # create folder if it doesn’t exist
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # copy image
    if os.path.exists(src_path):
        shutil.copy(src_path, os.path.join(dest_folder, img_name))

print("Finished organizing dataset")
```
