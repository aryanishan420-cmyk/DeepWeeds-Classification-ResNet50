```python
import pandas as pd
import os
import shutil
from tqdm import tqdm
```

```python
# folders
processed_folder = "../data/processed"
labels_folder = "../data/raw/labels"

train_folder = "../data/train"
val_folder = "../data/val"
test_folder = "../data/test"
```

```python
train_df = pd.read_csv(os.path.join(labels_folder, "train_subset0.csv"))
val_df = pd.read_csv(os.path.join(labels_folder, "val_subset0.csv"))
test_df = pd.read_csv(os.path.join(labels_folder, "test_subset0.csv"))

print("Train size:", len(train_df))
print("Val size:", len(val_df))
print("Test size:", len(test_df))
```

```python


labels_master = pd.read_csv("../data/raw/labels/labels.csv")

label_map = {str(int(r.Label)): r.Species for _, r in labels_master.iterrows()}

print("Label map loaded")
```

```python


processed_folder = "../data/processed"

def copy_images(df, target_folder):

    folders = os.listdir(processed_folder)

    for _, row in tqdm(df.iterrows(), total=len(df)):

        img = str(row["Filename"]).split(".")[0]
        label = str(int(row["Label"]))   # numeric label

        species = label_map.get(label, None)
        if species is None:
            continue

        # find matching processed folder
        match = None
        for f in folders:
            if species.lower() in f.lower():
                match = f
                break

        if match is None:
            continue

        src1 = os.path.join(processed_folder, match, img + ".jpg")
        src2 = os.path.join(processed_folder, match, img + ".JPG")

        dst = os.path.join(target_folder, match)
        os.makedirs(dst, exist_ok=True)

        if os.path.exists(src1):
            shutil.copy(src1, os.path.join(dst, img + ".jpg"))
        elif os.path.exists(src2):
            shutil.copy(src2, os.path.join(dst, img + ".JPG"))
```

```python
copy_images(train_df, "../data/train")
copy_images(val_df, "../data/val")
copy_images(test_df, "../data/test")
```
